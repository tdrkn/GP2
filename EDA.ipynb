```python
def dict_to_list(row):
    array = []
    for el in row:
        for key, value in el.items():
            if key == 'name':
                array.append(value)
    return '|'.join(array)
```

```python
def dict_to_list_for_companies(row):
    array = []
    for el in row:
        for key, value in el.items():
            if key == 'company':
                array.append(value['name'])
    return '|'.join(array)
```

```python
dict_to_list(df.loc[0, 'game_modes'])
```

```python
dict_to_list_for_companies(df.loc[0, 'involved_companies'])
```

```python
df['new_modes'] = df.loc[df['game_modes'].isna() == False, 'game_modes'].apply(dict_to_list)
df['new_genres'] = df.loc[df['genres'].isna() == False, 'genres'].apply(dict_to_list)
df['new_platforms'] = df.loc[df['platforms'].isna() == False, 'platforms'].apply(dict_to_list)
df['new_themes'] = df.loc[df['themes'].isna() == False, 'themes'].apply(dict_to_list)
df['new_companies'] = df.loc[df['involved_companies'].isna() == False, 'involved_companies'].apply(dict_to_list_for_companies)
df1 = df.drop(columns = ['game_modes', 'genres', 'involved_companies', 'platforms', 'themes', 'slug', 'external_games'])
```

```python
array = ['PC (Microsoft Windows)', 'Mac', 'Linux', 'PlayStation', 'Xbox', 'Android', 'iOS', 'Nintendo', 'Arcade', 'Wii']
for platform in array:
    df1[f'platform_{platform}'] = df1['new_platforms'].str.contains(platform, case = False, na=False, regex = False).astype(int)
df1['platform_Mobile'] = ((df1['platform_Android'] + df1['platform_iOS']) > 0).astype(int)
df2 = df1.drop(columns = ['platform_Android', 'platform_iOS'])
df2['platform_Other'] = ~df2[['platform_PC (Microsoft Windows)', 'platform_Mac', 'platform_Linux', 'platform_PlayStation', 'platform_Xbox', 'platform_Mobile', 'platform_Nintendo', 'platform_Arcade', 'platform_Wii']].any(axis = 1)
df2['platform_Other'] = df2['platform_Other'].astype(int)
```
